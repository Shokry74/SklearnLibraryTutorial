# Metrics in Sklearn:

Metrics calculate the cost function and lose function like (Mean Absolute Error , Mean Square Error , Meadian Absolute Error)
## in Regression:

#### Mean Absolute Error (MAE):

```python
from sklearn.metrics import mean_absolute_error
y_test = [3,-0.5,2,7]
y_pred = [2.5,0.0,2,8]


MAE = mean_absolute_error(y_test,y_pred ,multioutput = 'uniform_average' )  #(              )
MAE
```

```python
from sklearn.metrics import mean_absolute_error
y_test = [[3,-0.5],
        [2,7]]
y_pred = [[2.5,0.0]
          ,[2,8]]
 

MAE = mean_absolute_error(y_test,y_pred ,multioutput = 'raw_values')  #(              )
MAE
```

#### Mean Squared Error (MSE):

```python
from sklearn.metrics import mean_squared_error
y_test = [3,-0.5,2,7]
y_pred = [2.5,0.0,2,8]


MSE = mean_squared_error(y_test,y_pred ,multioutput = 'uniform_average' )  #(                )
MSE
```

#### Median Absolute Error (MAE):

```python
from sklearn.metrics import median_absolute_error
y_test = [3,-0.5,2,7]
y_pred = [2.5,0.0,2,8]


MAE = median_absolute_error(y_test,y_pred ,multioutput = 'uniform_average' )  #(                )
MAE
```

### Generic Univariate Select :

### E1 :

```python
from sklearn.feature_selection import GenericUnivariateSelect , chi2 , f_classif
from sklearn.datasets import load_breast_cancer
X , y = load_breast_cancer(return_X_y = True )
print(X.shape)
print("***************")
print(y.shape)
print("***************")
guselect = GenericUnivariateSelect(score_func = chi2 , mode = 'k_best' , param = 3)  
#    3     30    
newdata = guselect.fit(X,y)
# to know the columns that selected by model
print(newdata.get_support())

n_data = newdata.transform(X)
print(n_data.shape)
```
